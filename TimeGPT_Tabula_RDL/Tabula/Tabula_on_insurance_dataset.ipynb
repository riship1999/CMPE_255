```python
from tabula import Tabula
import pandas as pd
```

```python
data = pd.read_csv("/content/Tabula/Real_Datasets/Insurance_compressed.csv")
```

```python
categorical_columns = ["sex", "children", "sm", "region"]
model = Tabula(llm='distilgpt2', experiment_dir = "insurance_training", batch_size=32, epochs=10, categorical_columns = categorical_columns)
```

```python
model.fit(data)
```

```python
import torch
torch.save(model.model.state_dict(), "insurance_training/model_10epoch.pt")
```

```python
# synthetic_data = model.sample(n_samples=1338)
# synthetic_data.to_csv("insurance_10epoch.csv", index=False)
# Ensure the model is on CPU
device = torch.device("cpu")
model.model.to(device)

# Ensure sampling respects CPU
synthetic_data = model.sample(n_samples=50, device=device)  # Pass CPU device explicitly if sample allows
synthetic_data.to_csv("insurance_10epoch.csv", index=False)
```
