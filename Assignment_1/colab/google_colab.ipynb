```python
# LISTS --> Mutable, contains element of different datatypes, operations like append, insert, extend, delete are possible


grocery_list = ['Juice', 'Tomatoes', 'Potatoes', 'Bananas']
print('first item', grocery_list[1])
grocery_list[0] = "Green Juice"
print(grocery_list)
#slicing
print(grocery_list[1:3])

other_events = ['Wash Car', 'Pick up Kids', 'Cash Check']
to_do_list = [other_events, grocery_list]
print(to_do_list)


print(to_do_list[1][1])

#append
grocery_list.append('onions')
print(to_do_list)

#Insert at index
grocery_list.insert(1, "Pickle")

#Remove
grocery_list.remove("Pickle")

#Sorts items in list
grocery_list.sort()

#Reverse sort items in list
grocery_list.reverse()

#del deletes an item at specified index
del grocery_list[4]
print(to_do_list)

#We can combine lists with a +
to_do_list = other_events + grocery_list
print(to_do_list)

#Get length of list
print(len(to_do_list))

#Get the max item in list
print(max(to_do_list))

#Get the minimum item in list
print(min(to_do_list))
```

```python
# TUPLE --> Immutable, cannot be changed

# Tuple
tuple = (3, 1, 4, 1, 5, 9)

# Convert tuple into a list
new_list = list(tuple)
print(new_list)
```

```python
# DICTIONARY

super_villains = {'Fiddler' : 'Isaac Bowin',
                  'Captain Cold' : 'Leonard Snart',
                  'Weather Wizard' : 'Mark Mardon',
                  'Mirror Master' : 'Sam Scudder',
                  'Pied Piper' : 'Thomas Peterson'}

print(super_villains['Mirror Master'])

# Delete
del super_villains['Captain Cold']
print(super_villains)
# Replace
super_villains['Captain Cold'] = 'abc'
# Get the value for the passed key
print(super_villains.get("Pied Piper"))
# Get a list of dictionary keys
print(super_villains.keys())
# Get a list of dictionary values
print(super_villains.values())
```

```python
# OOPs

class Vehicle:
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year

  def display_info(self):
    print("Make:", self.make)
    print("Model:", self.model)
    print("Year:", self.year)


car1 = Vehicle("Toyota", "Corolla", 2022)
car2 = Vehicle("Honda", "Civic", 2023)


car1.display_info()
print()  # Print an empty line for separation
car2.display_info()
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
from PIL import Image
import io

def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
        async function takePhoto(quality) {
            const div = document.createElement('div');
            const video = document.createElement('video');
            const stream = await navigator.mediaDevices.getUserMedia({video: true});

            document.body.appendChild(div);
            div.appendChild(video);
            video.srcObject = stream;
            await video.play();

            // Resize the output to match video resolution.
            google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            canvas.getContext('2d').drawImage(video, 0, 0);
            stream.getTracks().forEach(track => track.stop());
            div.remove();

            const dataUrl = canvas.toDataURL('image/jpeg', quality);
            return dataUrl;
        }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

# Capture the photo
filename = take_photo()
print(f'Photo saved as {filename}')

# Load the captured image
image = Image.open(filename)

# Optionally, display the captured image using PIL
image.show()

# Convert to a format suitable for processing (e.g., OpenCV)
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img_array = np.array(image)
img_cv2 = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

# Display the image using cv2_imshow (for Colab)
cv2_imshow(img_cv2)

# Note: No need for cv2.waitKey(0) or cv2.destroyAllWindows() in Colab
```
