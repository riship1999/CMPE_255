```python
import pandas as pd
from pycaret.regression import *
import gradio as gr
```

```python
# Load dataset
from pycaret.datasets import get_data
data = get_data('boston')
```

```python
# Select a few important features for prediction
data = data[['rm', 'lstat', 'ptratio', 'medv']]  # Only keep relevant columns

# Initialize the setup
reg_experiment = setup(data, target='medv', session_id=123, verbose=True)

# Train the best model
best_model = compare_models()

# Define the predict function to accept individual feature inputs
def predict(RM, LSTAT, PTRATIO):
    # Create a single-row DataFrame with the selected inputs
    input_data = pd.DataFrame([[RM, LSTAT, PTRATIO]],
                              columns=['rm', 'lstat', 'ptratio'])

    # Make the prediction
    prediction = predict_model(best_model, data=input_data)

    print(prediction.columns)

    # Ensure that we retrieve the correct column for the predicted price
    predicted_value = prediction[prediction.columns[-1]].values[0]  # Select the last column

    # Return the predicted house price
    return float(predicted_value)

# Define Gradio interface with inputs for selected features
input_interface = [
    gr.Number(label='Average number of rooms per dwelling (RM)'),
    gr.Number(label='Percentage of lower status of the population (LSTAT)'),
    gr.Number(label='Pupil-teacher ratio by town (PTRATIO)'),
]

output_interface = gr.Textbox(label="Predicted Median Price")

# Launch the Gradio interface
gr.Interface(fn=predict, inputs=input_interface, outputs=output_interface).launch(share=True,debug=True)
```
