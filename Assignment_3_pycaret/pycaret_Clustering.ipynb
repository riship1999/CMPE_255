```python
import pandas as pd

# Load the dataset into a DataFrame
data = pd.read_csv('Mall_Customers.csv')  # Check the file name after unzipping
data.head()  # Verify the data
```

```python
from pycaret.clustering import *

# Set up the PyCaret environment
# Exclude the non-numeric columns (like 'CustomerID' and 'Gender') from clustering
clf1 = setup(data=data, numeric_features=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'], session_id=42)
```

```python
# Create a KMeans clustering model
kmeans_model = create_model('kmeans')  # 'kmeans' is the code for KMeans
```

```python
# Assign the cluster labels to the original data
clustered_data = assign_model(kmeans_model)

# Display the first few rows of the clustered data
clustered_data.head()
```

```python
# Visualize the clusters
plot_model(kmeans_model)
```

```python
# Save the clustering model for future use
save_model(kmeans_model, 'kmeans_model')
```
