```python
import pandas as pd
from pycaret.time_series import *
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('day.csv')

# Convert 'dteday' to datetime format
df['dteday'] = pd.to_datetime(df['dteday'])

# Set 'dteday' as the index and select 'cnt' as the target variable
df.set_index('dteday', inplace=True)
df = df[['cnt']]
```

```python
# Initialize PyCaret time series environment
exp = setup(data=df, target='cnt', session_id=123)
```

```python
# Directly create an ARIMA model
model = create_model('arima')
```

```python
# Forecast the next 7 days
future_forecast = predict_model(model, fh=7)

# Print the forecasted values
print(future_forecast)
```

```python
# Plot the forecast
plot_model(model, plot='forecast', data_kwargs={'fh': 7})
```
