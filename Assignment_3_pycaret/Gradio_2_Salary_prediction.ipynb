```python
import zipfile
with zipfile.ZipFile("salary.zip", 'r') as zip_ref:
    zip_ref.extractall("salary_data")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('salary_data/Salary.csv')

# Inspect the dataset
print(df.head())
```

```python
from pycaret.regression import *

# Initialize PyCaret for regression
reg = setup(data=df, target='Salary', verbose=True, session_id=123)

# Train a model (we'll use the best model)
model = compare_models()
```

```python
def predict_salary(YearsExperience):
    input_data = pd.DataFrame([[YearsExperience]], columns=['YearsExperience'])
    predictions = predict_model(model, data=input_data)
    return predictions['prediction_label'][0]
```

```python
import gradio as gr

# Set up Gradio interface
inputs = [
    gr.Number(label="Years of Experience")
]

outputs = gr.Textbox(label="Predicted Salary")

# Launch the Gradio app
gr.Interface(fn=predict_salary, inputs=inputs, outputs=outputs, title="Salary Prediction Model").launch(debug=True)
```
